```python
sns.histplot(data['Profit'],kde=False,bins=20)
```

```python
sns.boxplot(data['Discount'])
```

```python
for col in data.select_dtypes(include='object'):
    print(col)
    print(data[col].value_counts(normalize=True))
    print()
```

```python
corr = data.corr()
sns.heatmap(corr, cmap='crest', annot=True, fmt='.2f', linewidths=0.5)
```

```python
for col in data.select_dtypes(include=np.number):
    plt.figure(figsize=(8,6))
    sns.distplot(data[col])
    plt.title(col)
```

```python
sns.pairplot(data, diag_kind='hist')
```

```python
for col in data.select_dtypes(include=np.number):
    plt.figure(figsize=(8,6))
    sns.boxplot(x='Category', y=col, data=data)
```

```python
plt.figure(figsize=(10,8))
sns.countplot(x='Sub-Category', hue='Region', data=data)
```

```python
plt.figure(figsize=(8,8))
plt.plot(data.groupby('Region')['Sales'].sum())
plt.title('Sales by Region')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.show()
```
